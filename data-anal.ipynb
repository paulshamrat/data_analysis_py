# genomics data

```python
import pandas as pd

# Create a DataFrame with genomics data
data = pd.DataFrame({
    'Sample_ID': ['Sample1', 'Sample2', 'Sample3'],
    'Gene': ['GeneA', 'GeneB', 'GeneC'],
    'Expression_Level': [10.5, 8.2, 12.7],
    'Mutation_Status': ['Wildtype', 'Mutation', 'Wildtype']
})

# Save the DataFrame as a CSV file
data.to_csv('genomics_data.csv', index=False)
```

```python
# Display the first few rows of the DataFrame
print(data.head())

# Get information about the DataFrame, such as column names and data types
print(data.info())

# Generate summary statistics of the data
print(data.describe())
```

```python
# Select specific columns
subset = data[['Sample_ID', 'Expression_Level']]

# Filter rows based on conditions
filtered_data = data[data['Mutation_Status'] == 'Mutation']

# Sort data by a column
sorted_data = data.sort_values('Expression_Level', ascending=False)

# Group data by a column and calculate aggregates
grouped_data = data.groupby('Gene').mean()

# Create new columns based on existing columns
data['Expression_Category'] = pd.cut(data['Expression_Level'], bins=[0, 5, 10, 15], labels=['Low', 'Medium', 'High'])
```

```python
import matplotlib.pyplot as plt

# Bar plot of mutation status counts
mutation_counts = data['Mutation_Status'].value_counts()
plt.bar(mutation_counts.index, mutation_counts.values)
plt.xlabel('Mutation Status')
plt.ylabel('Count')
plt.title('Mutation Status Distribution')
plt.show()

# Scatter plot of expression levels
plt.scatter(data['Expression_Level'], data['Sample_ID'])
plt.xlabel('Expression Level')
plt.ylabel('Sample ID')
plt.title('Expression Level Distribution')
plt.show()
```

# large datasets

```python
import pandas as pd
import numpy as np

# Generate a large dataset with 1 million rows and 5 columns
n = 1000000
data = pd.DataFrame({
    'ID': np.arange(n),
    'Feature1': np.random.rand(n),
    'Feature2': np.random.randint(0, 10, n),
    'Feature3': np.random.choice(['A', 'B', 'C'], n),
    'Feature4': np.random.randn(n)
})

# 1. Exploring the Data
print(data.head())
print(data.info())
print(data.describe())

# 2. Data Manipulation
subset = data[['ID', 'Feature1']]
filtered_data = data[data['Feature2'] > 5]
sorted_data = data.sort_values('Feature1', ascending=False)
grouped_data = data.groupby('Feature3').mean()
data['Feature5'] = data['Feature1'] + data['Feature4']

# 3. Data Visualization
import matplotlib.pyplot as plt

plt.hist(data['Feature2'])
plt.xlabel('Feature2')
plt.ylabel('Count')
plt.title('Histogram of Feature2')
plt.show()

plt.scatter(data['Feature1'], data['Feature4'])
plt.xlabel('Feature1')
plt.ylabel('Feature4')
plt.title('Scatter Plot of Feature1 vs Feature4')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'C:/Users/paul/works/data_analysis_py/GSE225675_COUNTS_E11.5.txt'
df = pd.read_csv(file_path, delimiter='\t')

# Summary statistics
print(df.describe())

# Data distribution - Histogram for a specific column
plt.hist(df['female.control.1'], bins=10)
plt.xlabel('female.control.1')
plt.ylabel('Frequency')
plt.title('Histogram of female.control.1')
plt.show()

# Correlation analysis
correlation_matrix = df[['female.control.1', 'female.control.2', 'female.mutant.1']].corr()
print(correlation_matrix)

# Categorical analysis - Value counts for a specific column
feature_counts = df['female.control.2'].value_counts()
print(feature_counts)

# Grouping and aggregation
grouped_data = df.groupby('Feature3').mean()
print(grouped_data)

# Cross-tabulation analysis
cross_tab = pd.crosstab(df['female.control.3'], df['female.mutant.1'])
print(cross_tab)
```
